# Wordcloud presentation of Book content

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Let's make some word clouds!
# Terminal / Anaconda Prompt: conda install -c conda-forge wordcloud
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
from sklearn.feature_extraction import text 

custom_stopwords = ['shall','god','lord']
stop_words = text.ENGLISH_STOP_WORDS.union(custom_stopwords)
```

```python
# Let's make some word clouds!
# Terminal / Anaconda Prompt: conda install -c conda-forge wordcloud
from wordcloud import WordCloud

wc = WordCloud(stopwords=stop_words, background_color="white", colormap="Dark2",
               max_font_size=100, random_state=42)
```

```python
books_corpus = pd.read_pickle("books_corpus.pkl")
```

```python
# Reset the output dimensions
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [30, 20]

# Create subplots for each book
for index, book in enumerate(books_corpus.index):
    wc.generate(books_corpus['lemmatized_content'].loc[book])
    
    plt.subplot(3, 3, index+1)
    plt.imshow(wc, interpolation="bilinear")
    plt.axis("off")
    plt.title(book)
    
plt.show()
```

```python
from sklearn.feature_extraction import text 

custom_stopwords = ['shall','god','lord','book','say','said','hath','ye','copyright',
                    'true guru','people','thee','thou','unto','gur','guru','thy','rig']
stop_words = text.ENGLISH_STOP_WORDS.union(custom_stopwords)
```

```python
# Let's make some word clouds!
# Terminal / Anaconda Prompt: conda install -c conda-forge wordcloud
from wordcloud import WordCloud

wc = WordCloud(stopwords=stop_words, background_color="white", colormap="Dark2",
               max_font_size=100, random_state=42)
```

```python
# Reset the output dimensions
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [30, 20]

# Create subplots for each book's clean content.
# Non-english words eliminate from the corpus.
for index, book in enumerate(books_corpus.index):
    wc.generate(books_corpus['clean_content'].loc[book])
    
    plt.subplot(3, 3, index+1)
    plt.imshow(wc, interpolation="bilinear")
    plt.axis("off")
    plt.title(book)
    
plt.show()
```
